```python
import pandas as pd
```

```python
#pd.read_csv import the file into a variable called df
df = pd.read_csv('disney_plus_titles.csv')
```

```python
#head command shows the first 5 rows of you data
df.head()
```

```python
# selecting the specific column 'type'
df['type']
```

```python
#selecting the specific column show_id
df['show_id']
```

```python
#df.columns renames the columns to the given values in the brackets, in this case everything to uppercase
df.columns= ['Show ID', 'Type', 'Title', 'Director', 'Cast' ,'Country' , 'Date Added' , 'Release Year', 'Rating' , 'Duration', 'Listed In', 'Description']
```

```python
df
```

```python
# selects the specific string Movie inside the column Type and saves all values containing it in a variable called disney_movies
disney_movies = df[df['Type'].str.contains('Movie')]
disney_movies
```

```python
# selects the specific string TV inside the column Type and saves all values containing it in a variable called disney_shows

disney_shows = df[df['Type'].str.contains('TV')]
disney_shows
```

```python
#selects the first 10 elements in the newly created variable disney_movies and displays them
disney_movies['Type'].head(10)
```

```python
#selects the first 10 elements in the newly created variable disney_shows and displays them
disney_shows['Type'].head(10)
```

```python
#Gets all the Values TV-G contained within the column Rating and show how many of them are when stored in a new list
ratings_TV_G = df[df['Rating'].isin(['TV-G'])]
ratings_TV_G['Rating']
```

```python
#Gets all the Values PG contained within the column Rating and show how many of them are when stored in a new list
ratings_PG = df[df['Rating'].isin(['PG'])]
ratings_PG['Rating']
```

```python
#Gets all the Values TV-PG contained within the column Rating and show how many of them are when stored in a new list
ratings_TV_PG = df[df['Rating'].isin(['TV-PG'])]
ratings_TV_PG['Rating']
```
