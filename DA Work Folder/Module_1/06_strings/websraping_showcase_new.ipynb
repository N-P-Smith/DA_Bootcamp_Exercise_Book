```python
url = 'https://www.booking.com/searchresults.html?ss=London&ssne=London&ssne_untouched=London&label=gog235jc-1DCAEoggI46AdICVgDaFCIAQGYAQm4ARfIAQzYAQPoAQH4AQKIAgGoAgO4ArDuuaEGwAIB0gIkZmJhYjE4YzAtNDdhMy00MmY1LTk2NWItN2UzOTgyNTk1OWEx2AIE4AIB&aid=397594&lang=en-us&sb=1&src_elem=sb&src=searchresults&dest_id=-2601889&dest_type=city&checkin=2024-11-25&checkout=2024-03-26&ltfd=6%3A1%3A%3A%3A1&group_adults=2&no_rooms=1&group_children=0'
response  = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(response.text, 'html.parser') # creating a beautiful soup object
soup
```

```python
# Find all the hotel elements in the HTML document
hotels = soup.findAll('div', {'data-testid': 'property-card'})

pprint.pprint(hotels)
```

    Hint: Use a nice browser that allows you to easily view the HTML code side-by-side with the webpage. For example, in Google Chrome, you can right-click (or ctrl+click) a webpage and click ‘View Page Source’. You can also click ‘Inspect’ to get a more interactive comparison of the HTML code that corresponds to a section of the website you’re interested in.'

```python

hotels_data = []
# Loop over the hotel elements and extract the desired data
for hotel in hotels:
    # Extract the hotel name
    name_element = hotel.find('div', {'data-testid': 'title'})
    
    name = name_element.text.strip() if name_element else 'Not Available'
    
    # Extract the hotel location
    location_element = hotel.find('span', {'data-testid': 'address'})
    location = location_element.text.strip() if location_element else 'Not Available'
     
   

    # Extract the hotel score
    rating_element = hotel.find('div', {'data-testid': 'review-score'})
    score = rating_element.find('div', class_ = "a3b8729ab1 d86cee9b25")
    numbers = score.find('div', class_ = "ac4a7896c7")
    scores = [numbers.get_text(strip=True).split()[-1] for num in numbers]
   

    
    # Append hotes_data with info about hotel
    hotels_data.append({
        'name': name,
        'location': location,
        'rating': scores
    })
```

```python
hotels_data
```

```python
hotels_df = pd.DataFrame(hotels_data)
```

```python
hotels_df
```

```python
hotels_df['rating'] = hotels_df['rating'].str.get(0)
hotels_df
```
