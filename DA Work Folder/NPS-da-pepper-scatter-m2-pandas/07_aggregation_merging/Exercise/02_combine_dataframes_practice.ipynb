2. You have received some weather data (temperature) of the last year. For each month the average temperature was measured, only for a few months the maximum temperature could be measured. Anyway, you want to combine these two data without losing any information.

```python
weather_mean_data = pd.DataFrame({'Mean TemperatureF': [53.1, 70., 34.93548387, 28.71428571, 32.35483871, 72.87096774, 70.13333333, 35., 62.61290323, 39.8, 55.4516129 , 63.76666667],
                     'Month': ['Apr', 'Aug', 'Dec', 'Feb', 'Jan', 'Jul', 'Jun', 'Mar', 'May', 'Nov', 'Oct', 'Sep']})
weather_max_data = pd.DataFrame({'Max TemperatureF': [68, 89, 91, 84], 'Month': ['Jan', 'Apr', 'Jul', 'Oct']})
```

```python
weather_data = pd.merge(left = weather_mean_data, right = weather_max_data, how = "outer", on = "Month")

weather_data
```

```python
# Teacher solution.

import pandas as pd  

# Define the proper order of months 

month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'] 

# Convert the 'Month' column to a categorical type with the correct order 

weather_data['Month'] = pd.Categorical(weather_data['Month'], categories=month_order, ordered=True) 

# Sort the DataFrame by the 'Month' column 
 
weather_data_sorted = weather_data.sort_values('Month') 

# Display the sorted DataFrame
 
print(weather_data_sorted)
```

3. (Extra question: Can you fill in the average max. Temperature for the missing values in the Column `Max TemperatureF`)

```python
weather_data_sorted["Max TemperatureF"] = weather_data_sorted["Max TemperatureF"].fillna(0)

weather_data_sorted
```

```python
avg_max = weather_data_sorted["Max TemperatureF"].mean().round()

avg_max
```
