```python
muesli_delivery_dates = pd.merge(muesli_df_intern, muesli_df_campaign, how = "inner", on = "Order ID")

muesli_delivery_dates.drop_duplicates(inplace = True)

muesli_delivery_dates
```

```python
muesli_delivery_dates_drop = muesli_delivery_dates.drop(columns = ["Ready to Ship Date", "Customer Name"])

muesli_delivery_dates_drop
```

```python
muesli_delivery_dates_drop["Pickup Year"]=muesli_delivery_dates_drop["Pickup Date"].dt.year
muesli_delivery_dates_drop["Pickup Month"]=muesli_delivery_dates_drop["Pickup Date"].dt.month
muesli_delivery_dates_drop["Pickup Weekday"]=muesli_delivery_dates_drop["Pickup Date"].dt.dayofweek

muesli_delivery_dates_drop["Arrival Scan Year"]=muesli_delivery_dates_drop["Arrival Scan Date"].dt.year
muesli_delivery_dates_drop["Arrival Scan Month"]=muesli_delivery_dates_drop["Arrival Scan Date"].dt.month
muesli_delivery_dates_drop["Arrival Scan Weekday"]=muesli_delivery_dates_drop["Arrival Scan Date"].dt.dayofweek

muesli_delivery_dates_drop
```

```python
muesli_delivery_dates_drop["Number of days between Pickup and Arrival Scan"]=(muesli_delivery_dates_drop["Arrival Scan Date"] - muesli_delivery_dates_drop["Pickup Date"]).dt.days

muesli_delivery_dates_drop
```

```python
weekday_names = ["Monday", "Wednesday", "Friday"]

plt.subplots(figsize=(8,6))
sns.barplot(x = "Pickup Weekday", y = "Number of days between Pickup and Arrival Scan" , data = muesli_delivery_dates_drop).set(title = "Average Delivery Time by Weekday")
plt.xlabel("Weekday")
plt.ylabel("Delivery Timeframe in Days")
plt.xticks(ticks = range(3), labels = weekday_names);
```

```python
plt.subplots(figsize = (8,6))

sns.boxplot(y= "Number of days between Pickup and Arrival Scan", data = muesli_delivery_dates_drop).set(title = "Total Delivery Time")
plt.xlabel("Deliveries");
```
