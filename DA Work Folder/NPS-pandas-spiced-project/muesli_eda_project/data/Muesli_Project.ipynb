```python
muesli_df_campaign_drop = muesli_df_campaign.drop(columns = ["Customer Name"])

muesli_df_campaign_drop.head()
```

```python
muesli_df_process_drop = muesli_df_process.drop(columns = ["Row ID", "Ship Mode"])

muesli_df_process_drop.head()
```

```python
musesli_df_temp1 = pd.merge(muesli_df_process_drop, muesli_df_campaign_drop, how = "inner", on = "Order ID")

musesli_df_temp1.head()
```

```python
muesli_df_merged = pd.merge(musesli_df_temp1, muesli_df_intern, how = "inner", on = "Order ID")

muesli_df_merged.head()
```

```python
muesli_df_merged.drop_duplicates(inplace = True)

muesli_df_merged
```

```python
muesli_df_merged = muesli_df_merged[["Order ID", "Order Date", "Ready to Ship Date", "Pickup Date", "On Truck Scan Date", "Arrival Scan Date"]]

muesli_df_merged
```

```python
muesli_df_merged["Processing Time"] = (muesli_df_merged["Ready to Ship Date"] - muesli_df_merged["Order Date"]).dt.days
muesli_df_merged["Waiting Time"] = (muesli_df_merged["On Truck Scan Date"] - muesli_df_merged["Ready to Ship Date"]).dt.days
muesli_df_merged["Delivery Time"] = (muesli_df_merged["Arrival Scan Date"] - muesli_df_merged["On Truck Scan Date"]).dt.days

muesli_df_merged
```

```python
muesli_df_merged.set_index('Order ID', inplace = True)
muesli_df_merged
```

```python
muesli_df_new = muesli_df_merged[["Processing Time", "Waiting Time", "Delivery Time"]]
```

```python
fig, ax =plt.subplots(figsize=(10,6))

muesli_df_new.plot(kind='bar', stacked=True, ax=ax).set(title = "Service Overview Sample")
plt.ylabel("Total Days");
```

```python
plt.subplots(figsize=(8,6))
sns.lineplot(y = "Processing Time", x='Order Date', data = muesli_df_merged)
sns.lineplot(y = "Waiting Time", x='Order Date', data = muesli_df_merged)


```
