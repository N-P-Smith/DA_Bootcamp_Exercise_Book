### Let's work with gapminder dataset

```python
# plotly express has also some built-in datasets

df = px.data.gapminder()
```

```python
df
```

```python
# Let's create another dataframe which has only Germany

df_germany = df[df["country"] == "Germany"]
df_germany
```

```python
# Plot a line graph with plotly express which has markers and title

fig = px.line(data_frame = df_germany, x='year', y='lifeExp', height=300, title="Life Expectancy in Germany over the years", markers=True)
#fig.show(renderer = "browser")
fig.show()
```

#### Let's create a new dataframe only for the countries in Europe 

```python
df_europe = df[df["continent"] == "Europe"]
df_europe.head()
```

```python


fig = px.line(df_europe, x='year', y='lifeExp', color= "country")
    
fig.show(renderer = "browser")
```

### 2. Scatterplot

```python
# Let's continue with a scatterplot. This time, use size

fig = px.scatter(df_germany, 
                 x='year', 
                 y='lifeExp',
                 size = 'pop',
                 height=400, 
                 title="Life Expectancy in Germany")
fig.show(renderer = "browser")
```

#### Creating another dataframe and plotting a 3d scatterplot

```python
df_oceania = df[df["continent"] == "Oceania"]

fig = px.scatter_3d(df_oceania,
                    x='year',
                    y='gdpPercap',
                    z='pop',
                   color='country',
                   size='lifeExp')
fig.show(renderer = "browser")
fig.write_html('3d_scatter_plot.html')
```

#### Flashback

```python
fig = px.scatter(df, 
           x="gdpPercap", 
           y="lifeExp", 
           animation_frame="year", 
           size="pop", color="continent", hover_name="country",
           log_x=True, size_max=55, range_x=[100,100000], range_y=[25,90],)

fig.show(renderer = "browser")
fig.write_html('flashback.html')
```

### 3. Bar plot

```python
# Life expectancy in Germany over the years

fig = px.bar(df_germany , x='year', y='lifeExp', height=300)
    
    
fig.show(renderer = "browser")
    
fig.write_html("bar_simple.html")
```

```python
# comparing countries in a stacked barchart

countries = ['Germany', 'Belgium', 'Denmark']

fig = px.bar(df[df["country"].isin(countries)], 
             x='year', 
             y='lifeExp',
             barmode = "stack",  
             color='country',
             height=300)
    
fig.show(renderer = "browser")
fig.write_html("bar_stacked.html")
```

```python
# Changing the bar mode and orientation

fig = px.bar(df[df["country"].isin(countries)], 
             y='year', 
             x='lifeExp',  
             color='country',
             barmode='group', # modifying the barmode
             orientation="h",
             height=400)

fig.show(renderer = "browser")
fig.write_html("bar_side.html")
```

Type in `print(fig)` and you will see that a plotly graph is a large 
nested dictionary that contains all the visual declarations and the data for plotting.

`plotly.express` is a high level API that works similar to `seaborn` as columns 
of your DataFrame are mapped to visual parameters of the plot. But you can also directly change all values in the dictionary to fine-tune the plotly figure.

To do that use the functions `fig.update_*(key=dict(...))`.
